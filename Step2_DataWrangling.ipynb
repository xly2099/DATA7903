# Missing value handling

Calculates the proportion of missing values for each column. 

```python
percent_missing = df.isnull().sum() * 100 / len(df)
percentage = pd.DataFrame({'column_name': df.columns,
                                 'percent_missing': percent_missing})
percentage
```

For the PayloadMass column, there are 6.38% missing values. I used the mean interpolation method to process this column, and interpolated the mean value of this column of data to the missing position.

```python
PayloadMass_mean = df['PayloadMass'].mean()
PayloadMass_mean = round(PayloadMass_mean, 1)
df['PayloadMass'] = df['PayloadMass'].fillna(PayloadMass_mean)
```

For the LandingPad column, there are 31.91% missing values. Since the proportion of missing values in this column is relatively high, we create a new value missing and interpolate it.

```python
df['LandingPad'] = df['LandingPad'].fillna('missing')
```

For the Block column, there are 4.26% missing values, which is not very high. We create a new value of 0 and process it.

```python
df['Block'] = df['Block'].fillna('0')
```
