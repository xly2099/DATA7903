```python
import pandas as pd
import folium
```

```python
df = pd.read_csv('dataset_space_API_past_launches_raw.csv')

# Encode Outcome
df = df.replace({'Outcome': {'True Ocean': 1, 
                             'True RTLS': 1,
                             'True ASDS': 1,
                             'False Ocean':0,
                             'False RTLS':0,
                             'False ASDS':0,
                             'None ASDS':0,
                             'None None':0,},
                 'GridFins': {True: 1, False: 0},
                 'Reused': {True: 1, False: 0},
                 'Legs': {True: 1, False: 0}
                })
```

```python
# the Longtitude and Latitude
pd.DataFrame((df.loc[:, ['LaunchSite', 'Longitude', 'Latitude']])).head()
```

```python
success_rate = df.groupby('LaunchSite')['Outcome'].mean()

# Calculate the success rate of each LaunchSite.
CCSFSSLC40_success_rate = success_rate['CCSFS SLC 40']
KSCLC39A_success_rate = success_rate['KSC LC 39A']
VAFBSLC4E_success_rate = success_rate['VAFB SLC 4E']
KwajaleinAtoll_success_rate = success_rate['Kwajalein Atoll']

# Calculate the total number of each LaunchSite.
CCSFSSLC40_total = df['LaunchSite'].value_counts()['CCSFS SLC 40']
KwajaleinAtoll_total = df['LaunchSite'].value_counts()['Kwajalein Atoll']
KSCLC39A_total = df['LaunchSite'].value_counts()['KSC LC 39A']
VAFBSLC4E_total = df['LaunchSite'].value_counts()['VAFB SLC 4E']
```

```python
# Kwajalein Atoll

m = folium.Map(location=[9.047721, 167.743129], zoom_start=16)

popup_text = f'<b>Kwajalein Atoll</b><br>Success Rate: {KwajaleinAtoll_success_rate:.2%},<br>Number of launches: {KwajaleinAtoll_total}'

folium.Circle(
    radius = 100,
    location = [9.047721, 167.743129],
    popup = folium.Popup(popup_text, max_width= 200),
    color = 'red',
    fill = True
).add_to(m)

m
```

```python
# CCSFS SLC 40

m = folium.Map(location=[28.561857, -80.577366], zoom_start=16)

popup_text = f'<b>CCSFS SLC 40</b><br>Success Rate: {CCSFSSLC40_success_rate:.2%},<br>Number of launches: {CCSFSSLC40_total}'

folium.Circle(
    radius = 100,
    location = [28.561857, -80.577366],
    popup = folium.Popup(popup_text, max_width= 200),
    color = 'yellow',
    fill = True
).add_to(m)

m
```
